```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('jobs.db')
df = pd.read_sql('SELECT * FROM jobs', conn)
df.head()
```

```python
# 출처별 건수
df.groupby('source').size()
```

```python
# 원하는 SQL 쿼리 실행
pd.read_sql('''
    SELECT title, company, location, source
    FROM jobs
    WHERE source = 'linkedin'
    LIMIT 10
''', conn)
```

```python
# 양쪽 플랫폼에 동시 등록된 공고
pd.read_sql('''
    SELECT canonical_id, GROUP_CONCAT(source) AS sources, title, company
    FROM jobs
    GROUP BY canonical_id
    HAVING COUNT(DISTINCT source) > 1
''', conn)
```
